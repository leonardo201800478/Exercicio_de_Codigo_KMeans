Aqui, passaremos para uma classe de modelos de aprendizado de máquina não supervisionados: algoritmos de clustering.
Os algoritmos de agrupamento buscam aprender, a partir das propriedades dos dados, uma divisão ótima ou rotulagem discreta de grupos de pontos.

Muitos algoritmos de clustering estão disponíveis no Scikit-Learn e em outros lugares, mas talvez o mais simples de entender seja um algoritmo conhecido como *k-means clustering*, que é implementado em ``sklearn.cluster.KMeans``.

Começamos com as importações padrão:

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
```

## k-Means

O algoritmo *k*-means procura um número predeterminado de clusters em um conjunto de dados multidimensional não rotulado.
Ele faz isso usando uma concepção simples de como é o agrupamento ideal:

- O "centro do cluster" é a média aritmética de todos os pontos pertencentes ao cluster.
- Cada ponto está mais próximo de seu próprio centro de cluster do que de outros centros de cluster.

Essas duas suposições são a base do modelo k-means.
Em breve iremos mergulhar exatamente em *como* o algoritmo alcança essa solução, mas por enquanto vamos dar uma olhada em um conjunto de dados simples e ver o resultado *k*-means.

Primeiro, vamos gerar um conjunto de dados bidimensional contendo quatro blobs distintos.
Para enfatizar que este é um algoritmo não supervisionado, deixaremos os rótulos fora da visualização

```python

from sklearn.datasets import make_blobs
X, y_true = make_blobs(n_samples=3000, centers=4,
                       cluster_std=0.9, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50);
```

A olho nu, é relativamente fácil identificar os quatro grupos.


O algoritmo *k*-means faz isso automaticamente e, no Scikit-Learn, usa a típica API do estimador:

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

Vamos visualizar os resultados plotando os dados coloridos por esses rótulos. Também traçaremos os centros do cluster conforme determinado pelo estimador k-means:

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=8)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

A boa notícia é que o algoritmo *k*-means (pelo menos neste caso simples) atribui os pontos aos clusters de forma muito semelhante a como podemos atribuí-los a olho nu.
